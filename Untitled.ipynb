### Feature Engineering

**Date of Journey**

````python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
````

Alternate
```
df['Day'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

````python
df.head()
````

````python
## Converting Date, Month and year to Integer
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
````

````python
df.info()
````

````python
## droping the Date of journey column
df.drop('Date_of_Journey', axis = 1, inplace=True)
df.head()
````

**Arival Time**

````python
df['Arrival_Time'].str.split(' ').str[0]
````

````python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
````

````python
df.head(10)
````

````python
### Splitting the hours and min in arrival time
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]
````

````python
df.drop('Arrival_Time', axis=1, inplace=True)
````

````python
df.head()
````

````python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
````

````python
df.info()
````

**Departure Time**

````python
df['Dep_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_Min'] = df['Dep_Time'].apply(lambda x:x.split(':')[1])
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Min'] = df['Dep_Min'].astype(int)
````

````python
df.drop('Dep_Time', axis=1, inplace=True)
df.head()
````

````python
df.info()
````

**Total stops and Route**

````python
df['Total_Stops'].unique()
````

````python
df[df['Total_Stops'].isnull()]
````

````python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '2 stops': 2, '1 stop': 1, '3 stops': 3, 'nan': 2, '4 stops': 4})
df['Total_Stops']
````

````python
df.head(10)
````

````python
df.drop('Route', axis=1, inplace=True)
````

````python
df.loc[df['Duration'] == '5m']
````

````python
df.drop(6474, inplace=True)
df.drop(2660, inplace=True)
````

````python
df['hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['hours'] = df['hours'].astype('int')
````

````python
df['minutes'] =df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df['minutes'] = df['minutes'].fillna(0)
df['minutes'] = df['minutes'].astype('int')
````

````python
df['Duration'] = df['hours'] / 60  + df['minutes']
````

````python
df.drop('hours', axis=1, inplace=True)
df.drop('minutes', axis=1, inplace=True)
df.head()
````
